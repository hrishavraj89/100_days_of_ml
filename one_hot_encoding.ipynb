### OneHotEncoder using Sklearn

```python
#train-test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 0:4], df.iloc[:, -1], test_size = 0.2, random_state = 42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output = False, dtype = np.int32)
```

here, first we have to fetch the columns from the original DataFrame and them append the transformed Data into it

```python
x_train_new = ohe.fit_transform(x_train[['fuel', 'owner']])
```

```python
x_test_new = ohe.transform(x_test[['fuel', 'owner']])
```

### OneHotEncoding with top categories

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon')).astype(int).sample(5)
```
